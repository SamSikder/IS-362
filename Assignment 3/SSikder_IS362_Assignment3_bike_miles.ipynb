```python
import pandas as pd
```

```python
cumulative_miles = pd.Series(
    [55, 120, 180, 245, 310],
    index=["Day 1", "Day 2", "Day 3", "Day 4", "Day 5"]
)

cumulative_miles
```

```python
daily_miles = cumulative_miles.diff()
daily_miles.iloc[0] = cumulative_miles.iloc[0]

daily_miles
```

```python
miles_df = pd.DataFrame({
    "Cumulative Miles": cumulative_miles,
    "Miles Ridden Each Day": daily_miles
})

miles_df
```

### Bicycle Trip Mileage Analysis (Assignment 3)

This notebook examines bicycle mileage tracked over a period of several days. The collected odometer readings were stored in a pandas Series. By calculating the difference between successive cumulative values, the daily miles ridden were determined. The table above displays both the total mileage and the number of miles traveled each day.
