```python
names = ["LeBron", "Curry", "Durant"]
teams = ["Lakers", "Warriors", "Rockets"]

list(zip(names, teams))
```

## Using the zip() Function

The `zip()` function combines two lists by pairing elements that share the same index.
In this example, player names are paired with their corresponding teams.


```python
for name, team in zip(names, teams):
    print(name, "plays for the", team)
```

## What I Learned

- `zip()` allows you to combine two lists at once  
- It maintains the alignment of linked data  
- Loops can be applied to create clear, understandable code
