```python
import pandas as pd
```

## Dataset 1: Best-Selling Video Games

This dataset lists the best-selling video games of all time along with their sales and platforms.

```python
games = pd.read_csv("videogames.csv")
games.head()
```

```python
# Find the best selling game
games.sort_values(by="Sales_Millions", ascending=False)
```

```python
# Calculate average sales
games["Sales_Millions"].mean()
```

```python
# Count how many games per platform
games["Platform"].value_counts()
```

### Analysis

The dataset was loaded using pandas and examined using the `head()` function to preview the data.  
The data was sorted by sales to identify the best-selling video game.  
Additional analysis was performed to calculate the average sales and determine how many games appear on each platform.

## Dataset 2: World Series Champions

This dataset lists recent World Series champions along with the runner-up teams and number of games played.

```python
world_series = pd.read_csv("worldseries.csv")

world_series.head()
```

```python
# Count how many championships each team has
world_series["Champion"].value_counts()
```

### Analysis

The dataset was loaded using pandas and previewed using the `head()` function.  
The `value_counts()` function was used to determine how many times each team appears as a World Series champion.

## Dataset 3: Countries by GDP (Nominal)

This dataset lists several countries and their GDP values in trillions of U.S. dollars.

```python
gdp = pd.read_csv("gdp.csv")

gdp.head()
```

```python
# Sort countries by GDP
gdp.sort_values(by="GDP_Trillion_USD", ascending=False)
```

```python
# Calculate average GDP
gdp["GDP_Trillion_USD"].mean()
```

### Analysis

The GDP dataset was loaded using pandas and previewed using the `head()` function.  
The data was sorted to identify the countries with the highest GDP values.  
The average GDP was also calculated to understand the overall scale of the economies included in the dataset.
