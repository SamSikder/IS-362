This notebook analyzes movie ratings from multiple users using pandas.  
The dataset contains ratings for six movies from five users.  
The analysis calculates average ratings for each user and movie, then applies normalization to compare user preferences more fairly.

```python
import pandas as pd

df = pd.read_csv("movie_ratings.csv", index_col="User")

df
```

```python
df = df.drop(columns=["Unnamed: 7"])
df
```

```python
# Average rating given by each user
user_averages = df.mean(axis=1)

# Average rating received by each movie
movie_averages = df.mean(axis=0)

print("Average rating by user:")
print(user_averages)

print("\nAverage rating for each movie:")
print(movie_averages)
```

```python
# Normalize ratings by subtracting each user's average rating
normalized_df = df.sub(user_averages, axis=0)

print("Normalized Ratings:")
normalized_df
```

```python
# Average normalized ratings
normalized_user_avg = normalized_df.mean(axis=1)
normalized_movie_avg = normalized_df.mean(axis=0)

print("Average normalized rating by user:")
print(normalized_user_avg)

print("\nAverage normalized rating for each movie:")
print(normalized_movie_avg)
```

### Conclusion

Normalized ratings help account for differences in how users rate movies. 
Some users may rate movies more generously while others rate more strictly. 
By subtracting each user's average rating from their scores, we can better 
compare preferences between users.

One advantage of normalized ratings is that they remove individual bias 
from the data. This makes comparisons between users more fair and can 
improve recommendation systems.

One disadvantage is that normalized ratings are less intuitive than the 
original 1–5 scale. The normalized values are harder to interpret because 
they represent relative preference rather than the original rating.

```python
# Calculate the standard deviation of ratings for each user
user_std = df.std(axis=1)

# Standardize ratings: (rating - user mean) / user standard deviation
standardized_df = df.sub(user_averages, axis=0).div(user_std, axis=0)

print("Standardized Ratings:")
standardized_df
```

```python
# Average standardized ratings
standardized_user_avg = standardized_df.mean(axis=1)
standardized_movie_avg = standardized_df.mean(axis=0)

print("Average standardized rating by user:")
print(standardized_user_avg)

print("\nAverage standardized rating for each movie:")
print(standardized_movie_avg)
```

### Extra Credit: Standardized Ratings

Standardized ratings adjust the data by subtracting each user's average rating
and dividing by the standard deviation of that user's ratings. This accounts
for both the user's average rating and how spread out their ratings are.

Standardization can make comparisons between users more accurate because it
accounts for differences in rating behavior and rating variability.
